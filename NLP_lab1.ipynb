###### 1.1 слова, які є гапаксами (унікальні) і ті, що зустрічаються менше 5 разів

```python
# Вибираємо текст з NLTK
textN = text5
print(textN)
```

```python
# Частотний розподіл усіх слів у тексті
fdist = FreqDist(textN)
print(fdist)
```

```python
# 1.1 Знаходимо гапакси 
hapaxes = fdist.hapaxes()
print(hapaxes[:5])
```

```python
# Знаходимо слова, що трапляються менше ніж 5 разів
rare_words = [word for word in fdist if fdist[word] < 5]
print(rare_words[:5])
```

```python
# 2. Створюємо розподіл за довжиною слова
length_freq = FreqDist([len(word) for word in textN])

# Малюємо графік частоти залежно від довжини слова
plt.figure(figsize=(10,5))
length_freq.plot(title="Частота слів залежно від довжини слова")
```

```python
# 3. Знаходимо 10 найчастіших слів, що стоять в кінці речення
# шукаємо слова, після яких стоїть '.', '!', '?'
end_words = []
for i in range(len(textN) - 1):         
    if textN[i+1] in ['.', '!', '?']:
        end_words.append(textN[i])

end_fdist = FreqDist(end_words)
most_common_ends = end_fdist.most_common(10)


print("10 найчастіших слів, що стоять перед кінцем речення:")
for word, freq in most_common_ends:
    print(f"{word}: {freq}")
```

```python
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import wordpunct_tokenize
```

```python
with open("Intermezzo.txt", "r", encoding="utf-8") as f:
    raw = f.read()
    
# 2. Розбиваємо на слова/розділові знаки
tokens = wordpunct_tokenize(raw)

# 3. Створення nltk.Text
text = nltk.Text(tokens)
```

```python
# 4. Concordance — (вживання слова в контексті)
print("\nConcordance for 'я':")
text.concordance("я", width=60)
```

```python
# 5. Similar (пошук слів з подібним контекстом)
print("\nSimilar to 'я':")
text.similar("я")
```

```python
# 6. Візуалізація розподілу слів у тексті
text.dispersion_plot(["я", "втома", "сонце", "люди", "місто", "земля"])
```

```python
# 7. Частотний розподіл
fdist = nltk.FreqDist(text)
fdist
```

```python
# 8. Побудова графіка
fdist.plot(30, title="Найчастіші слова у 'Intermezzo'")

```

```python
# 9. Частотний розподіл 
print("\nНайпоширеніші слова:")
for word, freq in fdist.most_common(10):
    print(f"{word}: {freq}")
```
